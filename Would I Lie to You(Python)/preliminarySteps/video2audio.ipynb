##### `This notebook convert video files into audio files.`
Last updated on August 2nd, 2022

```python
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip
import moviepy.editor as mp
import os 
import pandas as pd

#a list of names
listOfName=os.listdir("Series 10")
try:
    listOfName.remove('.DS_Store')
except:
    pass

epID=[]
guestNames=[]
for filename in listOfName:
    #deal with the info
    parts=filename.split("- ")
    epName=parts[1][:-1]
    epID.append(epName)
    guestNames.append(parts[2][:-4])

    clip = mp.VideoFileClip("Series 10/"+filename)
    clip.audio.write_audiofile("audio/"+epName+".mp3")

#write the info into a csv file
df=pd.DataFrame({
    'EpisodeID': epID,
    'Names':guestNames
})
print(df)
df.to_csv("info/info10.csv")

```
